```python
import pandas as pd

# تعریف اسامی ستون ها به صورت دستی
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
                'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
                'hours-per-week', 'native-country', 'income']

# خواندن فایل CSV
data = pd.read_csv('adult.csv', header=None, names=column_names, na_values=" ?")

# نمایش ۵ ردیف اول برای اطمینان
print(data.head())
```

```python
data.info()
```

```python
# فقط روی ستون هایی که نوع داده Object (متنی) دارند اعمال شود.
for col in data.select_dtypes(include='object').columns:
    # حذف فضای خالی از ابتدا و انتهای تمام مقادیر
    data[col] = data[col].str.strip()

# بعد از تمیزکاری، دوباره وضعیت ستون 'income' را ببینیم.
print(data['income'].unique())
```

```python
# شمارش مقادیر گمشده (NaN) در هر ستون و تبدیل به درصد
missing_data_percentage = (data.isnull().sum() / len(data)) * 100
print(missing_data_percentage[missing_data_percentage > 0].sort_values(ascending=False))
```

```python
# 1. تفکیک متغیر هدف (Target)
y = data['income'].copy()

# 2. تفکیک ویژگی های ورودی (Features)
X = data.drop('income', axis=1)

# 3. ذخیره ستون جنسیت (ویژگی حساس) برای استفاده در Fairlearn
A = X['sex'].copy()

# 4. نمایش شکل (shape) هر کدام برای اطمینان
print(f"Shape of X (Features): {X.shape}")
print(f"Shape of y (Target): {y.shape}")
print(f"Shape of A (Sensitive Feature): {A.shape}")
```

```python
# لیستی از ستون هایی که باید حذف شوند
columns_to_drop = ['fnlwgt', 'education-num', 'native-country']

# حذف ستون های انتخابی از دیتافریم X
X = X.drop(columns_to_drop, axis=1)

# حالا ستون 'sex' که ویژگی حساس است را نیز باید از X حذف کنیم 
# (چون نباید مستقیماً در آموزش مدل استفاده شود)
X = X.drop('sex', axis=1) 

# نمایش شکل جدید برای اطمینان
print(f"New Shape of X (Features): {X.shape}")
```

```python
# تبدیل تمام ستون‌های متنی (Object) به ستون‌های عددی (One-Hot Encoding)
X_processed = pd.get_dummies(X)

# نمایش شکل جدید X
print(f"Shape of X after One-Hot Encoding: {X_processed.shape}")
```

```python
from sklearn.model_selection import train_test_split

# تقسیم داده‌ها به نسبت ۷۰٪ آموزش و ۳۰٪ آزمون
# random_state=42 تضمین می‌کند که هر بار تقسیم‌بندی شما یکسان باشد (اصلی مهم در MLOps)
X_train, X_test, y_train, y_test, A_train, A_test = train_test_split(
    X_processed, y, A, test_size=0.3, random_state=42
)

# نمایش ابعاد برای اطمینان (باید ۷۰٪ و ۳۰٪ از ۴۸۸۴۳ را نشان دهد)
print(f"Train set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

```python
# 1. تعریف تابع تبدیل
def to_binary(y):
    # ابتدا تمام فاصله های اضافی را از رشته ها حذف می کنیم (trim/strip)
    # سپس مقایسه را انجام می دهیم
    return (y.str.strip() == '>50K').astype(int) # استفاده از .str.strip() برای تمیزکاری

# 2. اعمال تبدیل روی متغیرهای هدف
y_train_binary = to_binary(y_train)
y_test_binary = to_binary(y_test)
```

```python
from sklearn.model_selection import train_test_split

# تقسیم داده‌ها
X_train, X_test, y_train, y_test, A_train, A_test = train_test_split(
    X_processed, y, A, test_size=0.3, random_state=42
)

# نمایش ابعاد برای اطمینان (حتماً باید در انتهای سلول باشد)
print(f"Train set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

```python
from sklearn.linear_model import LogisticRegression

# 1. ایجاد نمونه ای از مدل
baseline_model = LogisticRegression(solver='liblinear', random_state=42)

# 2. آموزش مدل بر روی داده های Train
# ********************** خط اصلاح شده **********************
# مدل باید بر روی y_train_binary (0 و 1) آموزش داده شود، نه y_train (متنی)
baseline_model.fit(X_train, y_train_binary) 

# 3. پیام موفقیت
print("Baseline Model (Logistic Regression) training complete on BINARY labels.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# 1. پیش بینی روی داده های Test
y_pred = baseline_model.predict(X_test)

# ********************** خطوط اصلاح شده **********************
# 2. محاسبه دقت (Accuracy) - استفاده از y_test_binary
accuracy = accuracy_score(y_test_binary, y_pred) 

# 3. نمایش گزارش طبقه بندی (Classification Report) - استفاده از y_test_binary
report = classification_report(y_test_binary, y_pred) 
# ************************************************************

# 4. نمایش خروجی ها
print(f"Overall Accuracy on Test Set: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(report)
```

```python
from fairlearn.metrics import MetricFrame, count
from sklearn.metrics import accuracy_score

# 1. تعریف یک دیکشنری از معیارهای مورد نظر
metrics = {
    "accuracy": accuracy_score,
    "sample_size": count  # برای دیدن تعداد نمونه در هر گروه
}

# 2. محاسبه معیارهای گروهی (استفاده از A_test برای تقسیم بندی)
grouped_metrics = MetricFrame(
    metrics=metrics, 
    y_true=y_test,          # پاسخ واقعی
    y_pred=y_pred,          # پیش بینی مدل
    sensitive_features=A_test # ستون حساس (جنسیت)
)

# 3. نمایش نتایج
print("Grouped Accuracy and Sample Size:")
print(grouped_metrics.overall)
print("\nAccuracy per Group (Sex):")
print(grouped_metrics.by_group)
```

```python
# Import تمام ماژول های مورد نیاز برای GridSearch
from fairlearn.reductions import GridSearch, EqualizedOdds
from sklearn.linear_model import LogisticRegression
```

```python
# 1. تعریف مدل پایه
estimator = LogisticRegression(solver='liblinear', random_state=42)
constraint = EqualizedOdds()
mitigation_model = GridSearch(
    estimator,
    constraints=constraint
)

# 2. آموزش مدل جدید (از y_train_binary استفاده کنید)
mitigation_model.fit(X_train, y_train_binary, sensitive_features=A_train)

print("Fairlearn GridSearch Model training complete.")
```

```python
# 1. پیش بینی روی داده های Test با مدل جدید (GridSearch)
# این متد، پیش بینی هر یک از مدل های کاندید را برمی گرداند
y_pred_mitigated = mitigation_model.predict(X_test)

# 2. نمایش شکل خروجی (باید تعداد ردیف های Test set و تعداد مدل های کاندید را نشان دهد)
print(f"Shape of mitigated predictions: {y_pred_mitigated.shape}")
```

```python
from sklearn.metrics import accuracy_score
from fairlearn.metrics import equal_opportunity_difference, selection_rate_difference

# 1. محاسبه دقت (Accuracy)
mitigated_accuracy = accuracy_score(y_test_binary, y_pred_mitigated)

# 2. محاسبه Equal Opportunity Difference (EOD)
mitigated_eod = equal_opportunity_difference(
    y_test_binary, 
    y_pred_mitigated, 
    sensitive_features=A_test
)

# 3. محاسبه Selection Rate Difference (SRD)
mitigated_srd = selection_rate_difference(
    y_test_binary, 
    y_pred_mitigated, 
    sensitive_features=A_test
)

# 4. نمایش مقایسه
print("--- Mitigation Model Performance Comparison ---")
print("Baseline Model Accuracy: 87.01%")
print("Baseline EOD: 0.6516") # داده‌های گزارش‌شده قبلی

print("\n--- Mitigated Model Results ---")
print(f"Mitigated Model Accuracy: {mitigated_accuracy * 100:.2f}%")
print(f"Mitigated Model EOD (Recall Disparity): {mitigated_eod:.4f}")
print(f"Mitigated Model SRD: {mitigated_srd:.4f}")
```
